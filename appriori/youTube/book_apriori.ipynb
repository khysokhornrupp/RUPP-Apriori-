```python
import numpy as np
```

```python
def load_tranc(path_data, order) :
    tranction = []
    with  open(path_data, 'r') as fid:
        for lines in fid:
            str_line = list(lines.strip().split(','))
            _t = list(np.unique(str_line))
            _t.sort(key=lambda x : order.index(x))  
            tranction.append(_t)
    return tranction
```
