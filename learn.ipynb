### Generator 

```python
def square_numbers(nums):
    for i in nums:
        yield i * i
nums = [1, 2, 3, 4, 5]

for i in square_numbers(nums):
    print(i)
```

```python
# change from [ ] to () for generator 
ny_nums = ( x * x for x in range(5) )
ny_nums
```

```python
for i in ny_nums:
    print(i)
```
