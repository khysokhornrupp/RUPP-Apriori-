### Generator 

```python
def square_numbers(nums):
    for i in nums:
        yield i * i
nums = [1, 2, 3, 4, 5]

for i in square_numbers(nums):
    print(i)
```

**F_k-1 x F_k-1**
</br>
Merge two frequent (k-1)-itemsets if their first (k-2) items are identical

```python
f = [['A','B', 'C'],['A','B','D'],['A','B','E'],['A','C','D'],['B','C','D'],['B','D','E'],['C','D','E']]
# f = [['A','B'],['A','C'],['B','C'],['B','D'],['C','D']]
# f = [  ['A', 'B'], ['B', 'C'], ['B', 'D'] ,['A', 'C'], ['C', 'D']]
# f = [  [ 'A', 'B', 'D' ], ['B', 'D', 'C'],  ['C', 'A', 'B'] ,['A', 'C', 'D']]
# f = [['A'],['B'],['C'],['D']]
k = len(f[1])
```

```python
prune_list = [  ]
```

```python
# candidate generation and caidate pruning 
merged_item = []
for i in range(len(f)):
    for j in range(i + 1, len(f)):
        if f[i][-(k-1):] == f[j][:k-1]:
            print(f"Itemset {f[i]} {f[j]}")
            item = set(f[i]).union(f[j])
            if len(prune_list) != 0:
                for p in prune_list:
                    if p != item:
                        merged_item.append(item)
            else :
                merged_item.append(item)
merged_item
```
