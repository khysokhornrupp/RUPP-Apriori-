```python
def polynomialofGFromAvgToB(x, a, b, avg, n=1):
    first_part = 1 / (2 * ((pow(b, n) - pow(avg, n))))

    second_part = (pow(b, n) - (2 * pow(avg, n))) / \
        2 * (pow(b, n) - pow(avg, n))

    result = first_part * pow(x, n) + second_part

    return  first_part, second_part


polynomialofGFromAvgToB(                                                                                                                                                                                                                                                     1, 0.1, 0.6, 0.2372)
```

```python
# m: real database size
def instance_selection(m, a=2, b=3):
    x0 = 5
    X = set()
    c = x0
    X.update({x0})
    while(len(X) != m):
        x_i = (a * c + b) % m
        if not set({x_i}).issubset(X):
            X.update({x_i})
        else:
            break
        c = x_i

    return X
```

```python
m = item_sets.shape[0]
ins = instance_selection(m=m)   
ins = np.fromiter(ins, int, len(ins)) 
ins
```
