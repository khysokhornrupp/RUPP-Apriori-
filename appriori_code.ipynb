```python
import pandas as pd
```

```python
def loadDataSet():
    return [[1, 3, 4], [2, 3, 5], [1, 2, 3, 5], [2, 5]]


def createC1(dataSet):
    C1 = []
    for transaction in dataSet:
        for item in transaction:
            if not [item] in C1:
                C1.append([item])
    C1.sort()
    return map(frozenset, C1)


def scanD(D, Ck, minSupport):
    ssCnt = {}
    for tid in D:
        for can in Ck:
            if can.issubset(tid):
                if not ssCnt.__contains__(can):
                    ssCnt[can] = 1
                else:
                    ssCnt[can] += 1
    numItems = float(len(D))
    retList = []
    supportData = {}
    for key in ssCnt:
        support = ssCnt[key] / numItems  # suppor of each item
        if support >= minSupport:
            retList.insert(0, key)
        supportData[key] = support

    return retList, supportData
```

```python
df = loadDataSet()
```

```python
c1 = createC1(df)
print(c1)
```

```python
D = [set(item) for item in df]
D
```

```python
l1, support0 = scanD(D, c1, 0.5)
```
