```python
def fib(n):
    if(n <= 1) : return n
    return  fib(n - 2) + fib(n - 1)
```

```python
def r_fib(n):
    if n <= 1:
        return n
    else:
        return (r_fib(n - 1) + r_fib(n - 2))
```

```python
def fibo(n, memo):
    if memo[n] != None:
        return memo[n]
    if n <= 1:
        return n
    else:
        res = fibo(n-1) + fibo(n+1)
        memo[n] = res
        return res  
def fib_memo(n):
    memo = [None] * (n + 1)
    return fibo(n, memo)
```
